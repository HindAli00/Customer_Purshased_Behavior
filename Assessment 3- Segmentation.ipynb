```python
X = df[['av_on_web', 'av_purch_price']]
# perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
# add cluster labels to customer data
df['Cluster'] = kmeans.labels_

# print segment details
for i in range(kmeans.n_clusters):
    segment_data = df[df['Cluster'] == i]
    print('Segment', i+1)
    print('Number of customers:', len(segment_data))
    print('Average on web:', segment_data['av_on_web'].mean())
    print('Average Purch:', segment_data['av_purch_price'].mean())
    print('\n')
```

```python
# Select the attributes to use for segmentation
attributes = ["av_on_web", "av_purch_price"]

# Normalize the data 
normalized_data = (df[attributes] - df[attributes].mean()) / df[attributes].std()

# Perform K-means clustering with k=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(normalized_data)

# Add the cluster labels to the dataframe
df["segment"] = kmeans.labels_

# Visualize the segments
plt.scatter(df["av_on_web"][df["segment"] == 0], df["av_purch_price"][df["segment"] == 0], color="green")
plt.scatter(df["av_on_web"][df["segment"] == 1], df["av_purch_price"][df["segment"] == 1], color="purple")
plt.scatter(df["av_on_web"][df["segment"] == 2], df["av_purch_price"][df["segment"] == 2], color="y")
plt.xlabel("Average time spent on product page")
plt.ylabel("Average purchase price")
print(df["segment"].mean())
plt.show()
```
